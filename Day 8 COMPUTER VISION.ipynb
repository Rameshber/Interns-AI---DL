```python
#PROBLEM ACC TO SIR.

img = np.ones((50,50,3),dtype = 'uint8') * 255
img[:25,:25,1:] = img[:25,:25,1:] * 0
img[:25,25:,::2] = img[:25,25:,::2] * 0
img[25:,:25,:2] = img[25:,:25,:2] * 0
img[25:,25:,2:] = img[25:,25:,2:] * 0

plt.imshow(img)

```

```python
#save the file using matplotlib
plt.imsave('squarebox1.png',img)
```

```python
#save the file using skimage
skimage.io.imsave('squarebox2.png',img)
```

```python
#save the file using opencv
cv2.imwrite('squarebox3.png',img)
```

```python
#IMP:- TRUE COLOR IMAGE OF CV2 IS NOT RGB BUT BGR.
# BUT TRUE COLOR OF ALL OTHER ARE AS:
# SKIMAGE    - TRUE COLOR IS RGB
# MATPLOTLIB - TRUE COLOR IS RGB
# PIL        - TRUE COLOR IS RGB
```

```python
#BUT TO GET THE IMAGE OF CV2 as RGB
cv2.imwrite('squarebox4.png',img[:,:,::-1])
```
