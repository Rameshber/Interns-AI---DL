```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
x1 = np.linspace(0,10,100)
fig = plt.figure()

plt.plot(x1, np.sin(x1),'-')
plt.plot(x1, np.cos(x1),'--')
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x1, np.sin(x1))
plt.show()

plt.subplot(2,1,2)
plt.plot(x1 , np.cos(x1))
plt.show()

plt.subplot(2,2,3)
plt.plot(x1, np.sin(x1))
plt.show()


```

```python
print(plt.gca())
print(plt.gcf())
```

```python
plt.plot([1,2,3,4])
plt.ylabel('Numbers')
plt.show()
```

```python
x = np.linspace(0,2,100)
plt.plot(x, x, label = 'Linear')
plt.plot(x, x**2, label = 'Quadratic')
plt.plot(x, x**3, label = 'Cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```
